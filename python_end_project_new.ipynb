```python
sns.lineplot(data=df,x='Age',y='Salary',marker='o',linestyle='--')
plt.title('AGE VS SALARY',fontsize=20,color='red')
plt.xlabel('Age',fontsize=15)
plt.ylabel('Salary',fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
import seaborn as sns
sns.barplot(data=df,x='Salary',y='Team',orient='h',color='c')
plt.title('SALARY VS TEAM',fontsize=20,color='red')
plt.xlabel('Team',fontsize=15,color='k')
plt.ylabel('Salary',fontsize=15,color='k')
plt.xticks(fontsize=15,color='k')
plt.yticks(fontsize=8,color='k')
plt.show()
```

```python
plt.hist(data=df,x='Salary',bins=10,color='m',edgecolor='k',alpha=0.7,histtype='bar',rwidth=0.8,label='Salary')
plt.title('SALARY DISTRIBUTION',fontsize=20,color='r',fontweight='bold')
plt.xlabel('Salary',fontsize=15,color='k')
plt.ylabel('Frequency',fontsize=15,color='k')
plt.legend()
plt.show()
```

```python
sns.boxplot(data=df,x='Salary',y='Position',orient='h',color='c',width=0.5,saturation=1,linewidth=1)
plt.title('SALARY VS POSITION',fontsize=20,color='r')
plt.xlabel('Salary',fontsize=15,color='k')
plt.ylabel('Position',fontsize=15,color='k')
plt.show()
```

```python
sns.boxplot(data=df,x='Height',y='Position',orient='h',color='g',width=0.5,saturation=1,linewidth=1)
plt.title('HEIGHT VS POSITION',fontsize=20,color='r')
plt.xlabel('Height',fontsize=15,color='k')
plt.ylabel('Position',fontsize=15,color='k')
plt.show()
```

```python
sns.lmplot(data=df,x='Salary',y='Age')
plt.title('SALARY VS AGE',fontsize=20,color='r')
plt.xlabel('Salary',fontsize=15,color='k')
plt.ylabel('Age',fontsize=15,color='k')
plt.show()
```
